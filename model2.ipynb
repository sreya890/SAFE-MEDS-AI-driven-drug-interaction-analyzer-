```python
import pandas as pd
from langchain_core.documents import Document
from langchain.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
import faiss
import numpy as np
import pickle

# Load dataset
file_path = "final_varied_drug_interaction.csv"
df = pd.read_csv(file_path)

# Convert dataset into document format
columns_to_include = list(df.columns)
documents = []

for i in range(len(df)):
    doc_text = f"Doctor Notes: {df.iloc[i]['doctor_notes']}\n"
    metadata = {col: str(df.iloc[i][col]) for col in columns_to_include if col != "doctor_notes"}
    documents.append(Document(page_content=doc_text, metadata=metadata))

# Create embeddings
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Create FAISS index
vector_store = FAISS.from_documents(documents, embedding_model)
vector_store.save_local("faiss_index")

# Save metadata separately
metadata_dict = {i: doc.metadata for i, doc in enumerate(documents)}
with open("faiss_index/index.pkl", "wb") as f:
    pickle.dump(metadata_dict, f)

print("FAISS Index created successfully!")
```
