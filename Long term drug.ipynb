```python
import pandas as pd
```

```python
data=pd.read_csv("FAERS_Dataset.csv")
data.head(5)
```

```python
data.isnull().sum()
```

```python
data=data.drop(['Patient ID','Time to Side Effect (Months)','Predicted Long-Term Risk'],axis =1)
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
encoder_drug = LabelEncoder()
encoder_condition = LabelEncoder()
encoder_effect = LabelEncoder()
```

```python
data['Drug Name'] = encoder_drug.fit_transform(data['Drug Name'])
data['Chronic Conditions'] = encoder_condition.fit_transform(data['Chronic Conditions'])
data['Side Effect'] = encoder_effect.fit_transform(data['Side Effect'])
```

```python
import joblib
joblib.dump(encoder_drug, "label_encoder_drug.pkl")
joblib.dump(encoder_condition, "label_encoder_condition.pkl")
joblib.dump(encoder_effect, "label_encoder_effect.pkl")
print("Encoders saved successfully!")
```

```python
data
```

```python
x= data.drop(columns=["Side Effect Severity"])
y = data["Side Effect Severity"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, stratify=y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_train, y_train)
```

```python
rf=RandomForestClassifier(random_state=42)
```

```python
params={'n_estimators':[50,80,100],
       'max_depth':[5,7,10],
       'min_samples_split':[2,5,10],
       'min_samples_leaf':[1,2,4],
       'bootstrap':[True,False]
       }
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
grid_search=GridSearchCV(param_grid=params,estimator=rf,cv=5,scoring='accuracy',verbose=2,n_jobs=-1)
```

```python
grid_search.fit(x_train,y_train)
```

```python
best_ref=grid_search.best_estimator_
```

```python
print(grid_search.best_params_)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
acc = accuracy_score(y_test, y_pred)
acc
```

```python
y_pred=best_ref.predict(x_test)
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```
