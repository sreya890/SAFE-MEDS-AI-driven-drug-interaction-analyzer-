```python
columns_to_include = list(df.columns)
data_lists = {col: df[col].astype(str).tolist() for col in columns_to_include} 
```

```python
documents = []
for i in range(len(df)):
    doc_text = f"Doctor Notes: {df.iloc[i]['doctor_notes']}\n"
    metadata = {col: df.iloc[i][col] for col in columns_to_include if col != "doctor_notes"}
    documents.append(Document(page_content=doc_text, metadata=metadata))
```

```python
import fitz  
import pymupdf
def extract_text_from_pdf(pdf_path):
    """Extract text from a PDF file."""
    doc =  pymupdf.open(pdf_path)
    text = " ".join([page.get_text("text") for page in doc])
    return text
```

```python
def retrieve_doctor_notes_from_pdf(pdf_path,top_k=1):
    """Retrieve relevant doctor notes based on extracted text from PDF."""
    extracted_text = extract_text_from_pdf(pdf_path)
    query_embedding = embedding_model.embed_query(extracted_text)  

   
    vector_store = FAISS.load_local("faiss_index", embedding_model, allow_dangerous_deserialization=True)

    search_results = vector_store.similarity_search_by_vector(query_embedding, k=top_k)

    retrieved_data = [
        {
            "doctor_notes": res.page_content,
            **res.metadata  
        }
        for res in search_results
    ]
    
    return retrieved_data
```
